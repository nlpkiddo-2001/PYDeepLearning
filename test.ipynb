```python
import sqlglot
import json

# Example 1: Simple SELECT query
json_sql = {
    "select": {
        "expressions": [
            {"column": {"this": {"identifier": "name"}}},
            {"column": {"this": {"identifier": "age"}}}
        ]
    },
    "from": {
        "expressions": [
            {"table": {"this": {"identifier": "users"}}}
        ]
    }
}

# Convert JSON to SQL string
expression = sqlglot.parse_one(json.dumps(json_sql), read="mysql")
sql_string = expression.sql()
print("Example 1 - Simple SELECT:")
print(sql_string)
print()

# # Example 2: More complex query with WHERE clause
# json_sql_complex = {
#     "select": {
#         "expressions": [
#             {"column": {"this": {"identifier": "product_name"}}},
#             {"column": {"this": {"identifier": "price"}}}
#         ]
#     },
#     "from": {
#         "expressions": [
#             {"table": {"this": {"identifier": "products"}}}
#         ]
#     },
#     "where": {
#         "gt": {
#             "this": {"column": {"this": {"identifier": "price"}}},
#             "expression": {"literal": {"this": "100", "is_string": False}}
#         }
#     }
# }
#
# expression = sqlglot.parse_one(json.dumps(json_sql_complex), read="json")
# sql_string = expression.sql()
# print("Example 2 - SELECT with WHERE:")
# print(sql_string)
# print()
#
# # Example 3: Using sqlglot to convert SQL to JSON and back
# original_sql = "SELECT id, name, email FROM customers WHERE status = 'active' ORDER BY name LIMIT 10"
#
# # SQL -> JSON
# parsed = sqlglot.parse_one(original_sql)
# json_representation = parsed.json()
# print("Example 3 - Original SQL:")
# print(original_sql)
# print("\nJSON representation:")
# print(json.dumps(json.loads(json_representation), indent=2))
#
# # JSON -> SQL
# expression_from_json = sqlglot.parse_one(json_representation, read="json")
# reconstructed_sql = expression_from_json.sql()
# print("\nReconstructed SQL:")
# print(reconstructed_sql)
# print()
#
# # Example 4: Converting to different SQL dialects
# print("Example 4 - Converting to different dialects:")
# json_rep = parsed.json()
# expr = sqlglot.parse_one(json_rep, read="json")
#
# print("PostgreSQL:", expr.sql(dialect="postgres"))
# print("MySQL:", expr.sql(dialect="mysql"))
# print("SQLite:", expr.sql(dialect="sqlite"))
```
